In the bellow I get test user rate vector and train user rate vector then return Norm of these vector.

For calculate norm I sum square all member of vector then get square sum of these number.



```
def  Norm(test_user_vote,train_user_vote):
    norm1=0
    norm2=0
    for i in range(len(test_user_vote)):
        norm1+=test_user_vote[i]**2
        norm2+=train_user_vote[i]**2

    return math.sqrt(norm1),math.sqrt(norm2)
```

Return Dot product result of two vector

For calculate dot producting I get multiply element in sam index then sum all them  

```
def  Dot(test_user_vote,train_user_vote):


    dot=0
    for i in range(len(test_user_vote)):
        dot+=test_user_vote[i]*train_user_vote[i]

    return dot
```

Find cosinus similarity between two user 




Cosinus similarity equations is :


![alt text](https://drive.google.com/uc?id=1F-Bl_Ze71dOUZ_wdF3NA855P3Qh6Qd0E)

I get test user and train user ids and the data where a store test users data and train user data. Then I find the movies both of them watch .When I get movies both of them watch I get rate of these user for these film. Finaly , With cosinus similarity I calculate how much these close each other. 

Return cosinus value for use it later.

```
def similarity(test_user,train_user,test_User_data,train_User_data):
    
    
    test_user_vote=[]
    train_user_vote=[]
    for a_film_data_test in test_User_data.get(test_user):
        for a_film_data_train in train_User_data.get(train_user):
            if a_film_data_test[0]==a_film_data_train[0]:
                    
                if a_film_data_test[1]!=-1:
                    
                    test_user_vote.append(a_film_data_test[1])
                    train_user_vote.append(a_film_data_train[1])

    cos_sim=0
    
    if len(test_user_vote)>0  and len(train_user_vote)>0:
            

           
        dot=Dot(test_user_vote,train_user_vote)
            

        norm1 ,norm2=Norm(test_user_vote,test_user_vote)
            
            
        cos_sim=dot/(norm1*norm2)
    
    return cos_sim
```
